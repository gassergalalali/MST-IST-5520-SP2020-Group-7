## Data Visualization

# 1)

```python
area_vars = ['SquareFootAreaoftheLot', 'SquareFootAreaofFinishedBasement', 'SquareFootAreaofHouse']
price_vars = ['SalesPrice', 'ContractPrice']
plt.figure(figsize=(20, 10))
y_vars = price_vars
x_vars = area_vars
for y_i, y_var in enumerate(y_vars):
    for x_i, x_var in enumerate(x_vars):
        ax=plt.subplot(len(y_vars), len(x_vars), y_i * len(x_vars) + x_i + 1)
        ax.scatter(dt[x_var], dt[y_var], marker='x')
        plt.xlabel(x_var)
        plt.ylabel(y_var)
plt.show()
```

# 2) Construction Method vs. Contract Price

```python
df2 = dt[['ConstructionMethod','ContractPrice']]

df2.head()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna()

df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=1.0, value= 'Modular')
df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=2.0, value= 'Panelized/Precut')
df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=3.0, value= 'Site_built')

df2['ConstructionMethod'].value_counts()
```

```python
# Generate an aggregated DataFrame to represent the mean of Contract Price
mean_contract_price_by_ConstMethod = df2.groupby('ConstructionMethod').mean()

mean_contract_price_by_ConstMethod
```

```python
ax = sns.barplot(x="ConstructionMethod", y="ContractPrice", data=df2, capsize=.2)
plt.title('Average Contract Price')
plt.ylabel('Contract Price ($)')
plt.show()
```

```python
mean_contract_price_by_ConstMethod['ContractPrice'].plot(kind='pie',
                                                  colors=['darkorange','forestgreen','steelblue'],
                                                  autopct='%.1f',
                                                  fontsize=15,
                                                  figsize=(6,4),
                                                  explode=(0.05, 0.05, 0.05),
                                                  shadow=True)
plt.ylabel('')
plt.title('Average Contract Price')
plt.show()
```
