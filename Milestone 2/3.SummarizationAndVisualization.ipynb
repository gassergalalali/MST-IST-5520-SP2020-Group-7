## 2) Sales price vs Areas Variables

```python
area_vars = ['SquareFootAreaoftheLot', 'SquareFootAreaofFinishedBasement', 'SquareFootAreaofHouse']
price_vars = ['SalesPrice']
plt.figure(figsize=(35, 10))
y_vars = price_vars
x_vars = area_vars
for y_i, y_var in enumerate(y_vars):
    for x_i, x_var in enumerate(x_vars):
        ax=plt.subplot(len(y_vars), len(x_vars), y_i * len(x_vars) + x_i + 1)
        ax.scatter(dt[x_var], dt[y_var], marker='x')
        plt.xlabel(x_var)
        plt.ylabel(y_var)
plt.show()
```

```python
area_vars = ['SquareFootAreaoftheLot', 'SquareFootAreaofFinishedBasement', 'SquareFootAreaofHouse']

for y_i in area_vars:
        sns.jointplot(dt[y_i], dt['SalesPrice'], marker='x')
plt.show()
```

## 3) Construction Method vs. Sales Price

```python
df2 = dt[['ConstructionMethod','SalesPrice']]

df2.head()
```

```python
df2['ConstructionMethod'].unique()
```

```python
import warnings
warnings.filterwarnings('ignore')

df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=1.0, value= 'Modular')
df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=2.0, value= 'Panelized/Precut')
df2['ConstructionMethod'] = df2['ConstructionMethod'].replace(to_replace=3.0, value= 'Site_built')

df2['ConstructionMethod'].value_counts()
```

```python
# Generate an aggregated DataFrame to represent the mean of Sales Price per Construction Method
mean_sales_price_by_ConstMethod = df2.groupby('ConstructionMethod').mean()

mean_sales_price_by_ConstMethod
```

```python
ax = sns.barplot(x="ConstructionMethod", y="SalesPrice", data=df2, capsize=.2)
plt.title('Average Sales Price per Construction Method')
plt.ylabel('Sales Price ($)')
plt.show()
```

```python
mean_sales_price_by_ConstMethod['SalesPrice'].plot(kind='pie',
                                                  colors=['darkorange','forestgreen','steelblue'],
                                                  autopct='%.1f',
                                                  fontsize=15,
                                                  figsize=(6,4),
                                                  explode=(0.05, 0.05, 0.05),
                                                  shadow=True)
plt.ylabel('')
plt.title('Average Sales Price per Construction Method ')
plt.show()
```

## 4) Sale Year vs. Sales Price

```python
dt['SaleYear']=dt['SaleDate']//100
dt.head()
```

```python
df1 = dt[['SaleYear','SalesPrice']].copy()
dfplot = pd.DataFrame(df1.groupby(['SaleYear'])['SalesPrice'].agg('sum'))
dfplot = dfplot.reset_index()
dfplot
```

```python
dfplot.plot.bar(x='SaleYear', y='SalesPrice', rot=90)
plt.title('Sum of Sale Values of Properties vs. Sale Year')
plt.ylabel('Sum Sales Value in 10^9')
plt.show()
```

```python
df2 = pd.crosstab(index=dt['SaleYear'],columns="SalesPrice")
df2 = df2.reset_index()
df2.plot.bar(x='SaleYear', y='SalesPrice', rot=90)
plt.title('Counts of Sale of Properties vs. Sale Year')
plt.ylabel('Sum Sales Count in Year')
plt.show()
```

##### As you can see in both figures above there is a major drop in values during the period of around 2008-2009. This is reasoned by the 2008 economic crisis

# 5) Detecting Outliers in Sales Price according to Boxplot Rules
# (Need to be Revised)

```python
df2['SalesPrice'].plot.box()
plt.show()
```

From the above boxplot, we notice that the 'Sales Price' variable has some outliers. According to the boxplot rule, outliers are data points that are out of the range [Q1 â€“ 1.5IQR, Q3 + 1.5IQR]. Now, let's use the boxplot rule to detect outliers from the dataset.

```python
# Calculate the first quantile
first_quantile = df2['SalesPrice'].quantile(0.25)

# Calculate the first quantile
third_quantile = df2['SalesPrice'].quantile(0.75)

# Calculate the interquantile range (IQR)
IQR = third_quantile - first_quantile
```

```python
# Outliers below the boxplot
df2[df2['SalesPrice']<first_quantile-1.5*IQR]
```

```python
# Outliers above the boxplot
df2[df2['SalesPrice']>first_quantile-1.5*IQR]
```
